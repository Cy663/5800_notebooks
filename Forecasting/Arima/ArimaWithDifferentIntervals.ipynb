```python
import pandas as pd
import numpy as np
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_absolute_error, mean_squared_error

# 1) Load rides and parse timestamps
rides = pd.read_csv('arima_features.csv', parse_dates=['started_at','ended_at'])

# Remove anything from the decimal-point onward in the timestamp strings
for col in ['started_at','ended_at']:
    rides[col] = rides[col].astype(str).str.replace(
        r'\.\d+',     # “.” followed by one or more digits
        '',           # replace with nothing
        regex=True
    )

# Now every entry is pure “YYYY-MM-DD HH:MM:SS”
# Convert to datetime with a fixed format
rides['started_at'] = pd.to_datetime(
    rides['started_at'],
    format="%Y-%m-%d %H:%M:%S"
)
rides['ended_at'] = pd.to_datetime(
    rides['ended_at'],
    format="%Y-%m-%d %H:%M:%S"
)

# pick one station to model (e.g. station 822)
station = 822

# 2) Define your time bins
freq = '20T'
t0 = rides['started_at'].min().floor(freq)
t1 = rides['ended_at'  ].max().ceil (freq)
time_index = pd.date_range(t0, t1, freq=freq)

# 3a) count out_flow for this station
#    group only rides that started at `station`
out = (
    rides[rides.start_id==station]
      .groupby(pd.Grouper(key='started_at', freq=freq))
      .size()
)
# 3b) count in_flow for this station
in_  = (
    rides[rides.end_id  ==station]
      .groupby(pd.Grouper(key='ended_at', freq=freq))
      .size()
)

# 4) assemble into one DataFrame
df = pd.DataFrame(index=time_index)
df['out_flow'] = out
df['in_flow' ] = in_
df = df.fillna(0).astype(int)
df['netflow'] = df['out_flow'] - df['in_flow']

# 5) train/test split (80/20)
split = int(len(df)*0.8)
train, test = df.iloc[:split], df.iloc[split:]

# 6) fit ARIMA on the netflow series
model = ARIMA(train['netflow'], order=(1,0,1)).fit()
forecast = model.forecast(steps=len(test))

# 7) compute metrics
y_true = test['netflow']
y_pred = forecast
mae  = mean_absolute_error(y_true, y_pred)
# rmse = mean_squared_error(y_true, y_pred, squared=False)
rmse = mean_squared_error(y_true, y_pred)
mape = np.mean(np.abs((y_true - y_pred) / (y_true + 1e-8))) * 100

print(f"Station {station} — 20min netflow forecast errors:")
print(f" MAE:  {mae:.3f}")
print(f" RMSE: {rmse:.3f}")
print(f" MAPE: {mape:.2f}%")
```

```python
import pandas as pd
import numpy as np

# 1) Collect all unique station IDs (both start and end)
all_stations = pd.unique(
    rides[['start_id','end_id']]
    .values
    .ravel()
)

# 2) Randomly sample 200 without replacement
#    (set random_state for reproducibility)
rng = np.random.default_rng(seed=42)
random_stations = rng.choice(all_stations, size=200, replace=False)

# 3) If you need it as a list
random_stations = list(random_stations)

print(f"Randomly selected {len(random_stations)} stations:", random_stations[:10], "…")
```

```python
import pandas as pd
import numpy as np
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_absolute_error, mean_squared_error


freqs = ['10min','15min','20min','30min','60min']

results = []
for station in random_stations:
    # floor/ceil with the same “min” unit
    for freq in freqs:
        t0 = rides['started_at'].min().floor(freq)
        t1 = rides['ended_at']  .max().ceil (freq)
        idx = pd.date_range(t0, t1, freq=freq)

        out = (rides[rides.start_id==station]
               .groupby(pd.Grouper(key='started_at', freq=freq))
               .size())

        inflow = (rides[rides.end_id==station]
                  .groupby(pd.Grouper(key='ended_at', freq=freq))
                  .size())

        df = pd.DataFrame({'out_flow': out, 'in_flow': inflow}, index=idx).fillna(0)
        df['netflow'] = df['out_flow'] - df['in_flow']

        split = int(len(df)*0.8)
        train, test = df.iloc[:split], df.iloc[split:]

        model = ARIMA(train['netflow'], order=(1,0,1)).fit()
        forecast = model.forecast(steps=len(test))

        y_true, y_pred = test['netflow'], forecast
        mae  = mean_absolute_error(y_true, y_pred)
        mse  = mean_squared_error(y_true, y_pred)
        rmse = np.sqrt(mse)
        mape = np.mean(np.abs((y_true - y_pred) / (y_true + 1e-8))) * 100

        results.append({
            'station':  station,
            'interval': freq,
            'MAE':      mae,
            'MAPE (%)': mape,
            'RMSE':     rmse
        })

metrics = pd.DataFrame(results)
pivot = metrics.pivot(index='station', columns='interval', values=['MAE','MAPE (%)','RMSE'])
print(pivot)
```
