```python
# Force pandas to parse mixed formats safely
df['started_at'] = pd.to_datetime(df['started_at'], format='mixed', errors='coerce')
df['ended_at'] = pd.to_datetime(df['ended_at'], format='mixed', errors='coerce')

# Drop rows with invalid dates (if any)
df = df.dropna(subset=['started_at', 'ended_at'])

# Keep only year-month-day-hour, remove minutes and seconds
df['started_at'] = df['started_at'].dt.floor('min')
df['ended_at'] = df['ended_at'].dt.floor('min')

# Check result
print(df[['started_at', 'ended_at']].head())
```

```python
# Convert datetime columns to proper datetime objects
df['started_at'] = pd.to_datetime(df['started_at'])
df['ended_at'] = pd.to_datetime(df['ended_at'])

# Set the desired frequency (e.g., '1H' for 1-hour intervals)
freq = 'h'

# Aggregate rides out (departures) per station using start_id for the chosen time interval
rides_out = (df.groupby(['start_id', pd.Grouper(key='started_at', freq=freq)])
               .size().reset_index(name='rides_out'))

# Aggregate rides in (arrivals) per station using end_id for the chosen time interval
rides_in = (df.groupby(['end_id', pd.Grouper(key='ended_at', freq=freq)])
              .size().reset_index(name='rides_in'))

# Rename columns for consistency: rename 'start_id' and 'end_id' to 'station_id'
rides_out.rename(columns={'start_id': 'station_id', 'started_at': 'time'}, inplace=True)
rides_in.rename(columns={'end_id': 'station_id', 'ended_at': 'time'}, inplace=True)

# Merge the two DataFrames, using an outer join to retain all observations
netflow = pd.merge(rides_in, rides_out, on=['station_id', 'time'], how='outer')

# Replace any missing ride counts with 0 and calculate netflow as rides_in minus rides_out
netflow['rides_in'] = netflow['rides_in'].fillna(0)
netflow['rides_out'] = netflow['rides_out'].fillna(0)
netflow['netflow'] = netflow['rides_in'] - netflow['rides_out']

# Sort the result by station and time for clarity
netflow.sort_values(by=['station_id', 'time'], inplace=True)

# Preview the netflow data for many stations
print(netflow.head(10))
```

```python
# Convert the 'time' column to datetime if it isn't already
netflow['time'] = pd.to_datetime(netflow['time'])

# Option 1: Select one station to forecast (here we choose the first station_id)
station_id = netflow['station_id'].unique()[0]
print("Selected station_id for forecast:", station_id)

# Filter the data for the selected station
station_data = netflow[netflow['station_id'] == station_id].copy()

# Set the time column as the index and ensure it's sorted
station_data.set_index('time', inplace=True)
station_data = station_data.sort_index()

# Make sure the data is at regular hourly frequency.
# This will fill in any missing hours with netflow=0.
station_ts = station_data['netflow'].asfreq('h').fillna(0)

# Preview the time series
print(station_ts.head(10))
```

```python
import numpy as np
from sklearn.metrics import mean_absolute_error

# MAE is already provided by sklearn, but here we simply wrap it.
def compute_mae(actual, predicted):
    return mean_absolute_error(actual, predicted)

# Define MAPE function
def compute_mape(actual, predicted):
    actual, predicted = np.array(actual), np.array(predicted)
    # To avoid division by zero, we add a small number to the denominator if needed.
    return np.mean(np.abs((actual - predicted) / (actual + 1e-10))) * 100
```

```python
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA

def forecast_arima(ts, horizon, order=(1,1,1)):
    """
    Forecast the time series 'ts' using ARIMA for a given forecast horizon.
    
    Parameters:
      - ts: The time series (pandas Series) indexed by datetime.
      - horizon: Forecast horizon (number of periods ahead; since ts is hourly, horizon=24 means 24 hours ahead).
      - order: The ARIMA model order as a tuple (p, d, q), defaulting to (1,1,1).
    
    Returns:
      - A tuple of (MAE, MAPE, forecast, test series).
    """
    # Split the data: use the last 'horizon' observations for testing.
    train = ts.iloc[:-horizon]
    test = ts.iloc[-horizon:]
    
    # Fit the ARIMA model on the training data
    model = ARIMA(train, order=order)
    model_fit = model.fit()
    
    # Forecast for the next 'horizon' steps
    forecast = model_fit.forecast(steps=horizon)
    
    # Evaluate forecast with MAE and MAPE
    mae = compute_mae(test, forecast)
    mape = compute_mape(test, forecast)
    
    # Plot the training data, the test data, and the forecasted values
    plt.figure(figsize=(10, 6))
    plt.plot(train.index, train, label='Train')
    plt.plot(test.index, test, label='Test', marker='o')
    plt.plot(test.index, forecast, label='Forecast', marker='o', linestyle='--')
    plt.xlabel('Time')
    plt.ylabel('Netflow (Rides In - Rides Out)')
    plt.title(f'ARIMA Forecast for {horizon}-hour horizon (Station: {station_id})\nMAE: {mae:.2f}, MAPE: {mape:.2f}%')
    plt.legend()
    plt.show()
    
    print(f"For a {horizon}-hour forecast:")
    print("MAE:", mae)
    print("MAPE:", mape, "%")
    
    return mae, mape, forecast, test
```

```python
# Define the list of forecast horizons (in hours)
forecast_horizons = {
    "1-hour": 1,
    "3-hour": 3,
    "24-hour": 24,
    "7-day": 168  # 7 days * 24 hours per day
}

# Create a dictionary to store evaluation metrics for each horizon
evaluation_results = {}

for label, horizon in forecast_horizons.items():
    print(f"\n===== Forecasting for {label} =====")
    mae, mape, forecast, test = forecast_arima(station_ts, horizon)
    evaluation_results[label] = {"MAE": mae, "MAPE": mape}
```
